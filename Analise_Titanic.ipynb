```python
# Passo 1 - Importar bibliotecas
# ============================================

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


# Configuração de estilo dos gráficos

sns.set(style="whitegrid")
```

```python
# Passo 2 - Carregar os dados
# ============================================

url = "https://drive.google.com/uc?id=11HptTxJbUMRG16xpC39fcliba_-Z_J9d"
df = pd.read_csv(url)

# Visualizar primeiras linhas

df.head()
```

```python
# Passo 3 - Explorar informações básicas
# ============================================


# Estrutura do dataset

df.info()

# Estatísticas descritivas

df.describe()

# Verificar valores nulos

df.isnull().sum()
```

```python
# Passo 4 - Análises simples
# ============================================


# Taxa de sobrevivência geral

print("Taxa de sobrevivência geral:")
print(df['Survived'].value_counts(normalize=True))

# Sobrevivência por sexo

print("\nTaxa de sobrevivência por sexo:")
print(df.groupby('Sex')['Survived'].mean())

# Sobrevivência por classe

print("\nTaxa de sobrevivência por classe:")
print(df.groupby('Pclass')['Survived'].mean())
```

```python
# Passo 5 - Gráficos
# ============================================


# Gráfico 1 - Sobrevivência por Sexo

plt.figure(figsize=(8,6))
sns.countplot(x='Sex', hue='Survived', data=df, palette="Set2")
plt.title("Sobrevivência por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Quantidade de Passageiros")
plt.legend(title="Sobrevivência", labels=["Não Sobreviveu", "Sobreviveu"])
plt.savefig("sobrevivencia_por_sexo.png")
plt.show()
```

```python
# Gráfico 2 - Sobrevivência por Classe


plt.figure(figsize=(8,6))
sns.countplot(x='Pclass', hue='Survived', data=df, palette="Set3")
plt.title("Sobrevivência por Classe")
plt.xlabel("Classe")
plt.ylabel("Quantidade de Passageiros")
plt.legend(title="Sobrevivência", labels=["Não Sobreviveu", "Sobreviveu"])
plt.savefig("sobrevivencia_por_classe.png")
plt.show()
```

```python
# Gráfico 3 - # Gráfico de sobrevivência geral

# Gráfico de sobrevivência geral (corrigido)
plt.figure(figsize=(6,6))
sns.countplot(x='Survived', hue='Survived', data=df, palette="pastel", legend=False)
plt.title("Taxa de Sobrevivência Geral")
plt.xlabel("Sobrevivência")
plt.ylabel("Quantidade de Passageiros")
plt.xticks([0,1], ["Não Sobreviveu", "Sobreviveu"])
plt.savefig("taxa_sobrevivencia_geral.png")
plt.show()

```
